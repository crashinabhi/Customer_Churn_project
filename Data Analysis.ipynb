## 1. Importing csv file for the first dataframe 

```python
# Reading the csv file to a dataframe
members_data = pd.read_csv(r'C:\Users\ancya\Customer_Churn_project\Data Set\Member Initial Details.csv')

# Extracting the head
members_data.head()
```

```python
# Dropping the columns with NAN values
members_data = members_data.dropna()

# Checking for any null(nan) value in the dataframe
members_data.isnull().values.any()
```

#### Dropping the columns that are not needed for final analysis

```python
members_data = members_data.drop(['City', 'Registered_Via', 'Registered_On'], axis = 1)

# Extracting the head
members_data.head()
```

## 2. Importing csv file for the second dataframe

```python
# Reading the csv file to a dataframe
transactions_data = pd.read_csv(r'C:\Users\ancya\Customer_Churn_project\Data Set\Transactions.csv')

# Extracting the head
transactions_data.head()
```

```python
# Dropping the columns with NAN values
transactions_data = transactions_data.dropna()

# Checking for any null(nan) value in the dataframe
transactions_data.isnull().values.any()
```

#### Dropping the columns that are not needed for final analysis

```python
transactions_data = transactions_data.drop(['Transaction_Date', 'Membership_Expire_Date'], axis = 1)

# Extracting the head
transactions_data.head()
```

```python
transactions_data = transactions_data.drop(['Plan_List_Price', 'Actual_Amount_Paid'], axis = 1)

# Extracting the head
transactions_data.head()
```

```python
transactions_data = transactions_data.drop(transactions_data.columns[1], axis=1)

# Extracting the head
transactions_data.head()
```

## 3. Importing csv file for the third dataframe

```python
# Reading the csv file to a dataframe
logs_data = pd.read_csv(r'C:\Users\ancya\Customer_Churn_project\Data Set\User Logs.csv')

# Extracting the head
logs_data.head()
```

```python
# Dropping the columns with NAN values
logs_data = logs_data.dropna()

# Checking for any null(nan) value in the dataframe
logs_data.isnull().values.any()
```

#### Dropping the columns that are not needed for final analysis

```python
logs_data = logs_data.drop(logs_data.columns[[1,3,4,5]], axis=1)

# Extracting the head
logs_data.head()
```
